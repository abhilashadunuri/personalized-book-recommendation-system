```python
ratings.rename(columns={
    "User-ID" : "user_id",
    "Book-Rating" :"rating"},inplace=True) 
```

```python
ratings['user_id'].value_counts()
```

```python
Need=ratings['user_id'].value_counts()>200
```

```python
Need
```

```python
Need[Need].shape
```

```python
y=Need[Need].index
```

```python
y
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
rat_books=ratings.merge(books,on="ISBN")
```

```python
rat_books.head()
```

```python
num_rating=rat_books.groupby('title')['rating'].count().reset_index()
```

```python
num_rating
```

```python
num_rating.rename(columns={
    'rating':'no_of_ratings'},inplace=True)
```

```python
num_rating.head()
```

```python
final=rat_books.merge(num_rating,on='title')
```

```python
final.head(2)
```

```python
final=final[final['no_of_ratings']>=50]
```

```python
final.sample(10)
```

```python
final.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final.shape
```

```python
book_pivot=final.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance,suggestion=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=5)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
book_names=book_pivot.index
```

```python
import pickle
pickle.dump(model,open('artifacts/model.pkl','wb'))
pickle.dump(book_names,open('artifacts/book_names.pkl','wb'))
pickle.dump(final,open('artifacts/final.pkl','wb'))
pickle.dump(book_pivot,open('artifacts/book_pivot.pkl','wb'))
```

```python
def recommend_book(book_name):
    book_id=np.where(book_pivot.index==book_name)[0][0]
    distance,suggestion=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
    for i in range(len(suggestion)):
        books=book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
```

```python
recommend_book('The Summons')
```
